```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Data loading
csv_file_path = '../data/processed/preprocessed_tweets.csv'
df = pd.read_csv(csv_file_path)

df.head()
```

```python
# Calculate tweet length
df['tweet_length'] = df['tweet'].apply(len)

# Filter the DataFrame for positive and negative sentiments
positive_tweets = df[df['sentiment'] == 1] # positive
negative_tweets = df[df['sentiment'] == 0] # negative

# Create the plot
fig, axes = plt.subplots(2, 1, figsize=(8, 10), sharex=True, sharey=True)

# Plotting positive sentiments
sns.histplot(positive_tweets['tweet_length'], bins=40, ax=axes[0], color='green', kde=True)
axes[0].set_title('Positive Sentiment Tweet Length Distribution')
axes[0].set_xlabel('Tweet Length')
axes[0].set_ylabel('Frequency')

# Plotting negative sentiments
sns.histplot(negative_tweets['tweet_length'], bins=40, ax=axes[1], color='red', kde=True)
axes[1].set_title('Negative Sentiment Tweet Length Distribution')
axes[1].set_xlabel('Tweet Length')
axes[1].set_ylabel('Frequency')

#plt.tight_layout()
plt.show()
```

```python
negative_tweet_text = " ".join(tweet for tweet in negative_tweets['tweet'])

# Generate word cloud
plt.figure(figsize = (10,10))
wc = WordCloud(max_words = 800 , width = 1200 , height = 600,
               collocations=False).generate(negative_tweet_text)
plt.axis('off')
plt.imshow(wc)
```

```python
positive_tweet_text = " ".join(tweet for tweet in positive_tweets['tweet'])

# Generate word cloud
plt.figure(figsize = (10,10))
wc = WordCloud(max_words = 800 , width = 1200 , height = 600,
               collocations=False).generate(positive_tweet_text)
plt.axis('off')
plt.imshow(wc)
```
